## 感知机的局限性
### 异或门
异或门也被成为**逻辑异或**电路
当$x_1$或$x_2$中的其中一个为1时，才会输出1（异或是拒绝其他的意思）
y=0.5-x

```python
import  numpy as np
import  matplotlib.pyplot as plt
ax=plt.gca()
#去掉边框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#指定x与y轴
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
#设置x与y轴的位置
ax.spines['bottom'].set_position(('data', 0))
#设置为居中显示:'center' 代替 ('data', 0)
ax.spines['left'].set_position(('data', 0))#设置为(0,0)显示
x = np.arange(-3,3,0.5)
y = 0.5 - x
plt.xlim(-2,2) # x轴的数值显示范围
plt.ylim(-2,2)

plt.fill_between(x, -3, y, facecolor='green', alpha=0.1)
plt.scatter(0, 1, s=100, c='c',marker='^')
plt.scatter(1, 1, s=100, c='c',marker='^')
plt.scatter(1, 0, s=100, c='c',marker='^')
plt.scatter(0, 0, s=100, c='c')
plt.plot(x,y)
plt.show()
```

绿色区域表示感知机输出0的区域，这个区域与或门的性质一致
如果使用异或门则无法表示

```python
import  numpy as np
import  matplotlib.pyplot as plt
ax=plt.gca()
#去掉边框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#指定x与y轴
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
#设置x与y轴的位置
ax.spines['bottom'].set_position(('data', 0))
#设置为居中显示:'center' 代替 ('data', 0)
ax.spines['left'].set_position(('data', 0))#设置为(0,0)显示
plt.xlim(-2,2) # x轴的数值显示范围
plt.ylim(-2,2)
plt.scatter(0, 1, s=100, c='c',marker='^')
plt.scatter(1, 1, s=100, c='c')
plt.scatter(1, 0, s=100, c='c',marker='^')
plt.scatter(0, 0, s=100, c='c')
plt.show()
```
